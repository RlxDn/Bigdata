# 04_BeautifulSoup 활용
* 멜론 사이트 크롤링하기
* 제목, 가수, 앨범명 추출

```python
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://www.melon.com/chart/index.htm'

header = {
  'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/90.0.4430.229 Whale/2.10.123.42 Safari/537.36'
}
res = requests.get(url, headers=header)
res
```

```python
soup = BeautifulSoup(res.text, 'html.parser')
soup
```

## 노래제목 추출

```python
songs = soup.select('.ellipsis.rank01')
title = []
for song in songs:
  title.append(song.find('a').get_text())
title
```

```python
songs = soup.select('.ellipsis.rank02')
singer = []

for song in songs:
  singer.append(song.find('a').get_text())
singer
```

```python
songs = soup.select('.ellipsis.rank03')
album = []

for song in songs:
  album.append(song.find('a').get_text())
album
```
